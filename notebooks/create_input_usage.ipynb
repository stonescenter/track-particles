## creating input with diferent events

```python
path = "event_2000_2020_train.csv"
event_files = []
tracks_final = pd.DataFrame()

for i in range(2000, 2030):
    event_files.append("event00000"+str(i))

for i in event_files:
    event_name = str(i) 
    output =  event_name +".csv"
    create_input("/data/trackMLDB/train/train_1/", event_name, 
                 output = output, 
                 sort = True,
                 cylindrical = False, silent = True)
    tracks = pd.read_csv(output)
    print('concatenating everything is working ok! ...')
    tracks_final = pd.concat([tracks_final, tracks], ignore_index=True)
    #sys.os.delete(output)

print('Data global saved at %s', path)
tracks_final.to_csv(path, index = False)    
```
